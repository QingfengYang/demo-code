```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
fig = plt.figure()
ax = plt.axes()

x = np.linspace(-10, 10, 1000)
ax.plot(x, (np.sign(1-x) + np.sign(1+x) - 1))
```
