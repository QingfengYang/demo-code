```python
import pandas as pd
data_path="/Users/yangqingfeng/tmp/tmp_data/tag_distribution/distribution_20181129_0424_44_32.032"
df=pd.read_csv(data_path, sep='=', names=['tagid', 'distr'])
tagDF=df[df['tagid'] != '\\N']
tag_int = tagDF['tagid'].astype('int')
tagDF_int=tagDF.assign(tagid = tag_int)
tagDF_int.head()
```

```python
def split_distr(kv):
    parts = kv.split(':')
    return (float(parts[0]), int(parts[1]))
```

```python
# 创作者标签
import numpy as np
import matplotlib.pyplot as plt

# xiuxiu content producer
xx_producer_df=tagDF_int[(tagDF_int['tagid']>=1120000000) & (tagDF_int['tagid']<1130000000)]
#xx_producer_df

xp_dict = {}
for index, row in xx_producer_df.iterrows():
    #print('tagid: %s, distr: %s' % (row['tagid'], row['distr']))
    distribution = row['distr']
    list_str = distribution.split('\t')
    n_list_str = list(map(lambda x: split_distr(x), list_str))
    for k, v in n_list_str:
        #print('key=%s, value=%s' %(k, v))
        if k not in xp_dict:
            xp_dict[k] = v
        else:
            xp_dict[k] = xp_dict[k] + v

plt.rcParams['savefig.dpi'] = 300
plt.rcParams['figure.dpi'] = 300
plt.figure(1)
plt.subplot(211)
plt.plot(xp_dict.keys(), xp_dict.values(), 'bo', markersize=2)
plt.show()
```
